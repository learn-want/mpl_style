在Matplotlib中，`plt`是顶层接口，用于创建图表，设置图表属性，`fig`是图表对象，用于对图表全局的设置，`ax`是子图对象。`fig`表示整个图表对象，`ax`表示图表中的一个子图对象。一般来说，当需要对整个图表进行设置时，使用`fig`对象；当需要对某个子图进行设置时，使用`ax`对象。

例如，当需要设置图表的大小、分辨率等全局属性时，可以使用`fig`对象；当需要设置某个子图的坐标轴、标题、标签等属性时，可以使用`ax`对象。在代码中，通常会先创建一个`fig`对象，然后再在`fig`对象上创建一个或多个`ax`对象，最后对`ax`对象进行设置。

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
 
# Dataset:
a = pd.DataFrame({ 'group' : np.repeat('A',500), 'value': np.random.normal(10, 5, 500) })
b = pd.DataFrame({ 'group' : np.repeat('B',500), 'value': np.random.normal(13, 1.2, 500) })
c = pd.DataFrame({ 'group' : np.repeat('B',500), 'value': np.random.normal(18, 1.2, 500) })
d = pd.DataFrame({ 'group' : np.repeat('C',20), 'value': np.random.normal(25, 4, 20) })
e = pd.DataFrame({ 'group' : np.repeat('D',100), 'value': np.random.uniform(12, size=100) })
df=pd.concat([a,b,c,d,e])

# boxplot
ax = sns.boxplot(x='group', y='value', data=df)
# add stripplot
ax = sns.stripplot(x='group', y='value', data=df, color="orange", jitter=0.2, size=2.5)

# add title
plt.title("Boxplot with jitter", loc="left")
ax.grid(zorder=100, lw =0.5, color = 'lightgray')
# show the graph
plt.figure(figsize=(6,3),dpi =300)
plt.show()
```

## Violin

```python
import numpy as np
import seaborn as sns
import pandas as pd
 
# Dataset:
a = pd.DataFrame({ 'group' : np.repeat('A',500), 'value': np.random.normal(10, 5, 500) })
b = pd.DataFrame({ 'group' : np.repeat('B',500), 'value': np.random.normal(13, 1.2, 500) })
c = pd.DataFrame({ 'group' : np.repeat('B',500), 'value': np.random.normal(18, 1.2, 500) })
d = pd.DataFrame({ 'group' : np.repeat('C',20), 'value': np.random.normal(25, 4, 20) })
e = pd.DataFrame({ 'group' : np.repeat('D',100), 'value': np.random.uniform(12, size=100) })
df=pd.concat([a,b,c,d,e])

# plot violin chart
ax=sns.violinplot( x='group', y='value', data=df)
ax.set_axisbelow(True)
# add title
plt.title("Violin plot", loc="left")
ax.grid(zorder=100, lw =0.5, color = 'lightgray')
# show the graph
plt.figure(figsize=(10,5),dpi =300)
plt.show()
```

```python
url = "https://raw.githubusercontent.com/jennybc/gapminder/master/data-raw/08_gap-every-five-years.tsv"
df = pd.read_csv(url, sep='\t')
# Calculate the average values for each continent
average_data = df.groupby('continent')[['gdpPercap', 'lifeExp', 'pop']].mean()

# Normalize the data for better visualization
normalized_data = (average_data - average_data.mean()) / average_data.std()

# Create parallel plot
plt.figure(figsize=(8, 6))
parallel_plot = sns.lineplot(data=normalized_data.transpose(),
                             dashes=False,
                             markers=True,
                             markersize=8,
                        )

# Add title
plt.title('Parallel Plot \nAverage GDP, Life Expectancy, and Population by Continent')

# Remove y-axis ticks and tick labels
plt.yticks([])

# Add legend
plt.legend(title='Continent',
           bbox_to_anchor=(1, 1), # Shift legend to the right
          )
parallel_plot.grid(zorder=100, lw =0.5, color = 'lightgray')
# parallel_plot.set_axisbelow(True)
# Display the chart
plt.show()
```
