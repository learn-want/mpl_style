# test plotly object

```python
# 生成随机数据
np.random.seed(0)
x = np.linspace(0, 1, 100)
y = np.random.randn(100) + 2 * x

coefficients = np.polyfit(x, y, 1)
p = np.poly1d(coefficients)
fit_line = p(x)

scatter_trace = go.Scatter(
    x=x,
    y=y,
    mode='markers',
    marker=dict(
        size=8,
        symbol='circle'
    ),
    name='Data Points'
)

line_trace = go.Scatter(
    x=x,
    y=fit_line,
    mode='lines',
    line=dict(
        width=2
    ),
    name='Regression Line'
)

line_trace1 = go.Scatter(
    x=x,
    y=fit_line+5,
    mode='lines',
    line=dict(
        width=2
    ),
    name='Regression Line1'
)
line_trace3 = go.Scatter(
    x=x,
    y=fit_line+9,
    mode='lines',
    line=dict(
        width=2
    ),
    name='Regression Line2'
)
line_trace4 = go.Scatter(
    x=x,
    y=fit_line+9,
    mode='lines',
    line=dict(
        width=2
    ),
    name='Regression Line2'
)
layout = go.Layout(
title='Test',
xaxis=dict(title='X'),
yaxis=dict(title='Y'),
showlegend=True)
fig = go.Figure(data=[scatter_trace, line_trace,line_trace1,line_trace3,line_trace4],layout=layout)

    # *****在每次绘制图形时应用绘图模板
# fig.update_layout(template='plotly_white+test')
fig.update_layout(template='simple_white+uzh_plotly')
# fig.update_yaxes(mirror=True)
# fig.update_xaxes(mirror=True)
fig.show()
```
